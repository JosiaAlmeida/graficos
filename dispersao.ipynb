```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv("./trees.csv")
base.head()
```

```python
#Criando grafico usado para comparar 2 variaveis numericas
plt.scatter(x=base.Girth, y=base.Volume, color='blue', facecolors='none', marker ='*')
plt.title("Arvore")
plt.ylabel("Circunferência")
plt.xlabel("Volume")
```

```python
#Criando grafico sem marker
plt.plot(base.Girth, base.Volume)
plt.title("Arvore")
plt.ylabel("Circunferência")
plt.xlabel("Volume")
```

```python
#
sns.regplot(x=base.Girth, y=base.Volume, data=base, x_jitter=0.3, fit_reg=False)
```

```python
#Separando dados na mesma tela
```

```python
plt.figure(1) # Definimos a figura
plt.subplot(2,2,1) # Linha, Coluna, Posição 
plt.scatter(base.Girth, base.Volume)
plt.subplot(2,2,2) # Linha, Coluna, Posição 
plt.scatter(base.Girth, base.Height)
plt.subplot(2,2,3) # Linha, Coluna, Posição 
plt.scatter(base.Girth, base.Volume, marker='*')
plt.subplot(2,2,4) # Linha, Coluna, Posição 
plt.hist(base.Volume)
```
